```python
from sklearn.datasets import make_moons, make_blobs
import matplotlib.pyplot as plt
```

```python
noisy_moon = make_moons(n_samples=100,noise = 0.1)
noisy_blob = make_blobs(n_samples=100,centers=2,
                             cluster_std=[0.8, 0.5],
                             random_state=0)
```

```python
plt.figure()
plt.scatter(noisy_moon[0][:,0],noisy_moon[0][:,1],c=noisy_moon[1])
plt.figure()
plt.scatter(noisy_blob[0][:,0],noisy_blob[0][:,1],c = noisy_blob[1])
```
